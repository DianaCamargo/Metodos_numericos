```python
print 'SOLUCION EJERCICIOS DE ALGUNOS EJERCICIOS DEL CHAPRA 2007'
import math as mt
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
a=-0.5
b=2.5
c=4.5
x1=(-b+mt.sqrt(b**(2)-(4*a*c)))/2*a
x2=-2*c/(b-mt.sqrt(b**(2)-(4*a*c)))
print '######Ejercicio 5.1 Chapra#####\n Metodo biseccion \n ###Diana Camargo###'
print 'f(x)= –0.5x2 + 2.5x + 4.5' 
print 'Error Tolerado=', es, '%'
print 'ERPA=Error Relativo Porcentual Aproximado'
print 'Las raices son: x1=',x1,'y x2=', x2,'\n Aproximaremos x2 por bisección a 3 iteracciones:'
print ("_______________________________________________________________________________")
print "|%-5s | %-12s |   %-9s  |   %-9s  | %-9s  | %-9s  |" %('Iteraccion','    xi  ','   xu','  xr ',' ERPA (%)','ERPV (%)')
xi=5.0
xu=10.0
xr=0.0
fxi=0.0
fxu=0.0
vver=x2#LA RAIZ Verdadera
i=1
erpa=0.0
while(i<4):
    val_ante=xr
    fxi=(a*xi**(2))+(b*xi)+c
    fxu=(a*xu**(2))+(b*xu)+c
    xr=(xi+xu)/2
    fxr=(a*xr**(2))+(b*xr)+c
    ev=vver-xr
    erpv=abs(ev/vver)*100
    erpa=abs((xr-val_ante)/(xr))*100
    
    print ("__________________________________________________________________________________")
    print"|    %-6d |    %-9f |    %-9f |    %-9f |  %-10f|  %-10f|" %(i,xi,xu,xr,erpa,erpv)
    if(fxi*fxr<0):
        xu=xr
    if(fxi*fxr>0):
        xi=xr
    if (erpa<es or fxi*fxr==0 ):
        break
    i=i+1
print '\nValor Aproximado= ',xr

```

```python
import math as mt
print '###########################################################\n'
print 'Ejercicio 5.13 chapra \n , calcule la masa m de modo que la velocidad sea v = 35 m/s en t = 9s. \n Utilice el método de la falsa posición para determinar m a un nivel de es = 0.1%'
print ("__________________________________________________________________________")
print "|%-5s | %-12s |   %-9s  |   %-9s  | %-9s |" %('Iteraccion','    mi  ','   mu','  mr ',' ERPA (%)')
es=0.1
g=9.8
v=35.0
c=15.0
t=9.0
mi=50
mu=70
mr=0.0
fxi=0.0
fxu=0.0
i=1
erpa=0.0
while(True):
    val_ante=mr
    fxi=((g*mi/c)*(1-(mt.e**(-(c*t)/mi))))-v
    fxu=((g*mu/c)*(1-(mt.e**(-(c*t)/mu))))-v
    mr=mu-(fxu*(mi-mu)/(fxi-fxu))
    fxr=((g*mr/c)*(1-(mt.e**(-(c*t)/mr))))-v
    erpa=abs((mr-val_ante)/(mr))*100
    print ("________________________________________________________________________")
    print"|    %-6d |    %-9f |    %-9f |    %-9f |  %-10f|" %(i,mi,mu,mr,erpa)
    if(fxi*fxr<0):
        mu=mr
    if(fxi*fxr>0):
        mi=mr
    if (fxi*fxr==0 or erpa<es):
        break
    i=i+1
print '\nValor de m Aproximado= ',mr
```
