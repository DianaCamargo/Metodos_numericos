```python
import math as mt
import matplotlib
import numpy as np
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
print '######Ejemplo 5.6 Chapra#####\n Metodo BISECCION \n ###Diana Camargo###'
print 'f(x)=e^-x   - x' 
print 'Error Tolerado=', es, '%'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("_______________________________________________________________________________")
print "|%-5s | %-12s |   %-9s  |   %-9s  | %-9s  | %-9s  |" %('Iteraccion','    xi  ','   xu','  xr ',' ERPA (%)','ERPV (%)')
xi=-0.5
xu=1
xr=0.0
fxi=0.0
fxu=0.0
vver=0.56714329  #LA RAIZ Verdadera
i=1
ii=0
erpa=0.0
u1=[]
ea1=[]
ever1=[]
while(True):
    val_ante=xr
    fxi=(mt.e**(-xi))-xi
    fxu=(mt.e**(-xu))-xu
    xr=(xi+xu)/2
    fxr=(mt.e**(-xr))-xr
    ev=vver-xr
    erpv=abs(ev/vver)*100
    erpa=abs((xr-val_ante)/(xr))*100
    
    print ("__________________________________________________________________________________")
    print"|    %-6d |    %-9f |    %-9f |    %-9f |  %-10f|  %-10f|" %(i,xi,xu,xr,erpa,erpv)
    u1.append(ii)
    ever1.append(erpv)
    if(fxi*fxr<0):
        xu=xr
    if(fxi*fxr>0):
        xi=xr
    if (erpa<es or fxi*fxr==0 ):
        break
    i=i+1
    ii=ii+1
print '\nValor Aproximado= ',xr  
```

```python
import math as mt
import matplotlib
import numpy as np
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
print '######Ejemplo 5.6 Chapra#####\n Metodo falsa posicion \n ###Diana Camargo###'
print 'f(x)=e^-x   - x'  
print 'Error Tolerado=', es, '%'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("_______________________________________________________________________________")
print "|%-5s | %-12s |   %-9s  |   %-9s  | %-9s  | %-9s  |" %('Iteraccion','    xi  ','   xu','  xr ',' ERPA (%)','ERPV (%)')
xi=0.5
xu=1
xr=0.0
fxi=0.0
fxu=0.0
vver=0.56714329  #LA RAIZ Verdadera
i=1
ii=0
erpa=0.0
val_ante=0.0
u2=[]
ea2=[]
ever2=[]
while(True):
    val_ante=xr
    fxi=(mt.e**(-xi))-xi
    fxu=(mt.e**(-xu))-xu
    xr=xu-(fxu*(xi-xu)/(fxi-fxu))
    fxr=(mt.e**(-xu))-xu
    erpa=abs((xr-val_ante)/(xr))*100
    ev=abs(vver-xr)
    erpv=abs(ev/vver)*100
    print ("__________________________________________________________________________________")
    print"|    %-6d |    %-9f |    %-9f |    %-9f |  %-10f|  %-10f|" %(i,xi,xu,xr,erpa,erpv)
    u2.append(ii)
    ever2.append(erpv)
    if(fxi*fxr<0):
        xu=xr
    if(fxi*fxr>0):
        xi=xr
    if (erpa<es or fxi*fxr==0):
        break
    i=i+1
    ii=ii+1
print '\nValor Aproximado= ',xr  
```

```python
import math as mt
import matplotlib
import numpy as np
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
print '######Ejemplo 5.6 Chapra#####\n Metodo PUNTO FIJO \n ###Diana Camargo###'
print 'f(x)=e^-x   - x' 
print 'Error Tolerado=', es, '%'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("_____________________________________________________")
print "|%-5s | %-12s | %-9s  | %-9s |" %('Iteraccion','    xi+1  ',' ERPA (%)','ERPV (%)')

#Despejando x
#xi+1=(e^-xi)

ximas1=0.0
vver=0.56714329  #LA RAIZ Verdadera
i=1
ii=0
erpa=0.0
u3=[]
ea3=[]
ever3=[]
while(True):
    xi=ximas1
    ximas1=(mt.e**(-xi))
    erpa=abs((ximas1-xi)/(ximas1))*100
    ev=abs(vver-ximas1)
    erpv=abs(ev/vver)*100
    print ("____________________________________________________")
    print"|    %-6d |   %-9f |  %-10f|  %-10f|" %(i,ximas1,erpa,erpv)
    u3.append(ii)
    ever3.append(erpv)
    if (erpa<es):
        break
    i=i+1
    ii=ii+1
print '\nValor Aproximado= ',ximas1
```

```python
import math as mt
import matplotlib
import numpy as np
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
print '######Ejemplo 5.6 Chapra#####\n Metodo NEWTON-RAPHSON \n ###Diana Camargo###'
print 'f(x)=e^-x   - x' 
print 'Error Tolerado=', es, '%'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("_____________________________________________________")
print "|%-5s | %-12s | %-9s  | %-9s |" %('Iteraccion','    xi+1  ',' ERPA (%)','ERPV (%)')

#Hallando la derivada de la función
#f'(x)=(-e^-x) -1

ximas1=0.0
vver=0.56714329  #LA RAIZ Verdadera
i=1
ii=0
erpa=0.0
u4=[]
ea4=[]
ever4=[]
while(True):
    xi=ximas1
    ximas1=xi-(((mt.e**(-xi))-xi)/((-mt.e**(-xi))-1))
    erpa=abs((ximas1-xi)/(ximas1))*100
    ev=abs(vver-ximas1)
    erpv=abs(ev/vver)*100
    print ("____________________________________________________")
    print"|    %-6d |   %-9s |  %-10f|  %-10s|" %(i,ximas1,erpa,erpv)
    u4.append(ii)
    ever4.append(erpv)
    if (erpa<es):
        break
    i=i+1
    ii=ii+1
print '\nValor Aproximado= ',ximas1
```

```python
import math as mt
import matplotlib
import numpy as np
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
print '######Ejemplo 5.6 Chapra#####\n Metodo SECANTE \n ###Diana Camargo###'
print 'f(x)=e^-x   - x' 
print 'Error Tolerado=', es, '%'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("________________________________________________________________")
print "|%-5s | %-15s | %-10s  | %-12s |" %('Iteraccion','    xi+1  ',' ERPA (%)','ERPV (%)')
ximenos1=0.0
xi=1
ximas1=0.0
fximas1=0.0
fximenos1=0.0
fxi=0.0
vver=0.56714329  #LA RAIZ Verdadera
i=1
ii=0
erpa=0.0
u5=[]
ea5=[]
ever5=[]
while(True):
    fximenos1=(mt.e**(-ximenos1))-ximenos1
    fxi=(mt.e**(-xi))-xi
    ximas1=xi-((fxi*(ximenos1-xi))/(fximenos1-fxi))
    erpa=abs((ximas1-xi)/(ximas1))*100
    ev=abs(vver-ximas1)
    erpv=abs(ev/vver)*100
    print ("______________________________________________________________")
    print"|    %-6d |   %-9s |  %-10f|  %-10s|" %(i,ximas1,erpa,erpv)
    u5.append(ii)
    ever5.append(erpv)
    if (erpa<es):
        break
    i=i+1
    ii=ii+1
    xi=ximas1
print '\nValor Aproximado= ',ximas1
```

```python
import math as mt
import matplotlib
import numpy as np
n=5
es=0.5*(10**(2-n))*100 #Error tolerado
print '######Ejemplo 5.6 Chapra#####\n Metodo SECANTE MODIFICADA \n ###Diana Camargo###'
print 'f(x)=e^-x   - x' 
print 'Error Tolerado=', es, '%'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("________________________________________________________________")
print "|%-5s | %-15s | %-10s  | %-12s |" %('Iteraccion','    xi+1  ',' ERPA (%)','ERPV (%)')
delta=0.01
xi=1
fxi=0.0
vver=0.56714329  #LA RAIZ Verdadera
i=1
ii=0
erpa=0.0
u6=[]
ea6=[]
ever6=[]
while(True):
    xiMasDelta=xi+delta
    fxiMasDelta=(mt.e**(-xiMasDelta))-xiMasDelta
    fxi=(mt.e**(-xi))-xi
    ximas1=xi-((delta*fxi)/(fxiMasDelta-fxi))
    erpa=abs((ximas1-xi)/(ximas1))*100
    ev=abs(vver-ximas1)
    erpv=abs(ev/vver)*100
    print ("______________________________________________________________")
    print"|    %-6d |   %-9s |  %-10f|  %-10s|" %(i,ximas1,erpa,erpv)
    u6.append(ii)
    ever6.append(erpv)
    if (erpa<es):
        break
    i=i+1
    ii=ii+1
    xi=ximas1
print '\nValor Aproximado= ',ximas1
```
