```python
import matplotlib.pyplot as plt
import math as mt
import numpy
n=3
es=0.5*(10**(2-n))*100
print '######Ejercicio 3.5 Chapra#####\n Diana Camargo'
print 'f(x)=e^5' 
print 'Error Tolerado=', es, '%'
print 'ERPV=Error Relativo Porcentual Verdadero'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("________________________________________________________")
print "|%-5s | %-9s  | %-9s  | %-10s  |" %('Iteraccion','e**x aprox ','   ERPV','   ERPA')

x=5
resu=0.0
nresu=1.0
vver=0.0
vver=mt.e**(x)
i=1
erpa=0.0
y=[]
u=[]
ever=[]
ea=[]
while True:
    vresu=nresu
    resu=(x**(i))/mt.factorial(i)
    nresu= resu+nresu
    ev=vver-nresu
    erpv=(ev*100)/nresu
    erpa=(nresu-vresu)*100/nresu
    print ("_______________________________________________________")
    print"|    %-6d |    %-9f |  %-9f |   %-10f|" %(i,nresu,erpv,erpa)
    y.append(nresu)
    u.append(i)
    ever.append(erpv)
    ea.append(erpa)
    i=i+1 
    
    if (erpa<es):
        break

###################grafica primer funcion################
plt.plot(u,y,'r-', label='e^5')
plt.plot(u,ever,'b--', label='error verd')
plt.plot(u,ea,'g--', label='error aprox')
plt.title("SERIE MCLAURIN")
plt.xlabel("x")
#plt.xlim(0.9, 3)
plt.ylabel("f(x)")
#plt.ylim(0, 2)
plt.legend(loc=1)
plt.show()
print '\nValor Aproximado= ',nresu 

print '############################################'
print 'f(x)=1/e^5' 
print 'Error Tolerado=', es, '%'
print 'ERPV=Error Relativo Porcentual Verdadero'
print 'ERPA=Error Relativo Porcentual Aproximado'
print ("________________________________________________________")
print "|%-5s | %-9s  | %-9s  | %-10s  |" %('Iteraccion','e**x aprox ','   ERPV','   ERPA')
resu=0.0
nresu=1.0
vresu=0.0
vver=0.0
vver=1/mt.e**(x)
print 'dfghjklñ... ',vver
i=1
erpa=0.0
y2=[]
u2=[]
ever2=[]
ea2=[]
erpv=0.0
erpa=0.0
while True:
    vresu=nresu
    resu=(x**(i))/mt.factorial(i)
    nresu=resu+nresu
    nresu=1/nresu
    ev=vver-nresu
    erpv=(ev*100)/nresu
    erpa=((nresu-vresu)/nresu)*100
    print ("_______________________________________________________")
    print"|    %-6d |    %-9f |  %-9f |   %-10f|" %(i,nresu,erpv,erpa)
    y2.append(nresu)
    u2.append(i)
    ever2.append(erpv)
    ea2.append(erpa)
    i=i+1 
    
    if (erpa<es):
        break 
###############Grafica de la segunda función#########################
plt.plot(u2,y2,'r-', label='1/e^5')
plt.plot(u2,ever2,'b--', label='error verd')
plt.plot(u2,ea2,'g--', label='error aprox')
plt.title("SERIE MCLAURIN")
plt.xlabel("x")
#plt.xlim(0.9, 3)
plt.ylabel("f(x)")
#plt.ylim(0, 2)
plt.legend(loc=1)
plt.show()
print '\nValor Aproximado= ',nresu 
```
